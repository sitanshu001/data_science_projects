## **BAR CHARTS**

- bar charts are representation of frequency tables
- it is under matplotlib package
- the method name is **bar**
  - plt.bar

- step-1 read the data

step-2 apply value counts

step-3 value counts will give keys and values 

step-4 consider keys as x-axis 

step-5 consider values as y-axis

step-6 pass those two directly in pit.bar

```python
keys=visa_df['continent'].value_counts().keys()
values=visa_df['continent'].value_counts().values
plt.bar(keys,values)
```

```python
keys=visa_df['continent'].value_counts().keys()
values=visa_df['continent'].value_counts().values
plt.figure(figsize=(10,5))
plt.bar(keys,values)
plt.xlabel('class') # x-axis name 
plt.ylabel('class frequency') # y-axis name
plt.title('continent barchart')
plt.savefig('continent_barchart.jpg')
plt.show() # show always at last
```

## **subplots**

- we can plot the many plots in one layout
- suppose we have 25 plots want to show
- if we show one by one, you keep on scroll down
- that is not a good idea
- we can show all the plots using **subplot** method in a single layout
- subplot method available in matplot package

```python
plt.subplot(2,2,1)
plt.subplot(2,2,2)
plt.subplot(2,2,3)
plt.subplot(2,2,4)
```

```python
r=3
c=3
for i in range(1,r*c+1):
    plt.subplot(r,c,i)
```

```python
r=3
c=3
for i in range(1,r*c+2):
    plt.subplot(r,c,i)
```

```python
keys=visa_df['continent'].value_counts().keys()
values=visa_df['continent'].value_counts().values
plt.pie(values,labels=keys)

plt.show()
```

```python
keys=visa_df['continent'].value_counts().keys()
values=visa_df['continent'].value_counts().values
plt.pie(values,
        labels=keys,
        autopct='%0.2f%%')

plt.show()
```
